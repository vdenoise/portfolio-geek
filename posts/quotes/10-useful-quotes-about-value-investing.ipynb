### Value Investing


Value investing is an art that combines patience, wisdom, and a deep understanding of financial fundamentals. This investment philosophy, popularised by legends like Warren Buffett and Benjamin Graham, focuses on identifying undervalued companies with solid fundamentals. Here are ten insightful quotes from renowned investors, each accompanied by a brief explanation to help you grasp the essence of value investing.
:::

::: {.content-visible when-profile="website_fr"}
### Bienvenue dans l'approche "Value"

L'investissement _value_ est un art qui combine patience, sagesse et une compréhension approfondie des fondamentaux financiers. Cette philosophie d'investissement, popularisée par des légendes comme Warren Buffett et Benjamin Graham, consiste à identifier des entreprises sous-évaluées avec de solides fondamentaux. Voici dix citations perspicaces de célèbres investisseurs, chacune accompagnée d'une brève explication pour vous aider à saisir l'essence de l'investissement _value_.
:::

::: {layout-ncol=2}

:::{#first-column}

> "**Price is what you pay. Value is what you get.**"
> -- <cite>Warren Buffett</cite>

::: {.content-visible when-profile="website_en"}
This quote by Warren Buffett, one of the most successful investors of all time, encapsulates the core principle of value investing. It reminds investors that the market price of a stock is not always indicative of its intrinsic value. True value investors seek to purchase stocks that are priced below their intrinsic value, ensuring they get more than what they pay for in the long run.
:::


::: {.content-visible when-profile="website_fr"}
Cette citation de Warren Buffett, l'un des investisseurs les plus célèbres de tous les temps, résume le principe fondamental de l'investissement _value_. Elle rappelle aux investisseurs que le prix d'une action sur le marché n'est pas toujours représentatif de sa valeur intrinsèque. Les vrais investisseurs _value_ cherchent à acheter des actions dont le prix est inférieur à leur valeur réelle, s'assurant ainsi d'obtenir plus que ce qu'ils paient à long terme.
:::

> "**The stock market is filled with individuals who know the price of everything, but the value of nothing.**"
> -- <cite>Philip Fisher</cite>

::: {.content-visible when-profile="website_en"}
Philip Fisher’s quote highlights a common pitfall in investing—focusing solely on price without considering the underlying value of an asset. Value investors understand that a low price doesn't necessarily mean a good deal if the asset lacks fundamental strength. By concentrating on value, rather than just price, investors can make more informed decisions.
:::

::: {.content-visible when-profile="website_fr"}
La citation de Philip Fisher met en lumière un piège courant dans l'investissement : se concentrer uniquement sur le prix sans considérer la valeur sous-jacente d'un actif. Les investisseurs _value_ comprennent qu'un prix bas ne signifie pas nécessairement une bonne affaire si l'actif manque de solidité fondamentale. En se concentrant sur la valeur plutôt que sur le prix, les investisseurs peuvent prendre des décisions plus éclairées.
:::

> "**The individual investor should act consistently as an investor and not as a speculator.**"
> -- <cite>Benjamin Graham</cite>

::: {.content-visible when-profile="website_en"}
Benjamin Graham, the father of value investing, draws a clear distinction between investing and speculating. Speculators often chase short-term gains, driven by market trends or emotions. In contrast, value investors take a long-term approach, basing their decisions on careful analysis of a company's fundamentals and intrinsic value, rather than on market fluctuations.
:::

::: {.content-visible when-profile="website_fr"}
Benjamin Graham, le père de l'investissement _value_, établit une distinction claire entre investir et spéculer. Les spéculateurs poursuivent souvent des gains à court terme, guidés par les tendances du marché ou les émotions. En revanche, les investisseurs en valeur adoptent une approche à long terme, fondant leurs décisions sur une analyse minutieuse des fondamentaux et de la valeur intrinsèque d'une entreprise, plutôt que sur les fluctuations du marché.
:::
---

> "**In the short run, the market is a voting machine but in the long run, it is a weighing machine.**"
> -- <cite>Benjamin Graham</cite>

::: {.content-visible when-profile="website_en"}
Graham’s analogy illustrates the difference between short-term market behaviour and long-term value recognition. In the short term, stock prices are influenced by market sentiment and speculation. However, over time, the market tends to reflect the true value of a company, which is why value investors focus on long-term potential rather than immediate price movements.
:::

::: {.content-visible when-profile="website_fr"}
L'analogie de Graham illustre la différence entre le comportement du marché à court terme et la reconnaissance de la valeur à long terme. À court terme, les prix des actions sont influencés par le sentiment du marché et la spéculation. Cependant, avec le temps, le marché tend à refléter la vraie valeur d'une entreprise, c'est pourquoi les investisseurs en valeur se concentrent sur le potentiel à long terme plutôt que sur les mouvements de prix immédiats.
:::
---

> "**It’s far better to buy a wonderful company at a fair price than a fair company at a wonderful price.**"
> -- <cite>Warren Buffett</cite>

::: {.content-visible when-profile="website_en"}
Buffett’s wisdom suggests that quality trumps price when it comes to investing. A wonderful company with strong fundamentals and a durable competitive advantage will likely deliver better returns over time, even if purchased at a fair price, than a mediocre company bought at a discount. This principle encourages investors to prioritise the strength and potential of a business over short-term bargains.
:::

::: {.content-visible when-profile="website_fr"}
L'expérience de Buffett suggère que la qualité l'emporte sur le prix en matière d'investissement. Une entreprise merveilleuse avec de solides fondamentaux et un avantage concurrentiel durable offrira probablement de meilleurs rendements sur le long terme, même si elle est achetée à un prix juste, qu'une entreprise médiocre achetée à rabais. Ce principe encourage les investisseurs à privilégier la force et le potentiel d'une entreprise plutôt que les bonnes affaires à court terme.
:::

---

> "**Investing is most intelligent when it is most businesslike.**"
> -- <cite>Benjamin Graham</cite>

::: {.content-visible when-profile="website_en"}
Graham emphasises the importance of treating investing as a business rather than a game. Just as a business owner would carefully analyse every aspect of their operations, an intelligent investor should thoroughly research and understand the companies they invest in. This disciplined approach reduces the likelihood of emotional decisions and increases the chances of long-term success.
:::

::: {.content-visible when-profile="website_fr"}
Graham souligne l'importance de traiter l'investissement comme un _business_ plutôt qu'un jeu de hasaed. Tout comme un chef d'entreprise analyserait soigneusement chaque aspect de ses opérations, un investisseur intelligent devrait rechercher et comprendre à fond les entreprises dans lesquelles il investit. Cette approche disciplinée réduit la probabilité de décisions émotionnelles et augmente les chances de succès à long terme.
:::

---

> "**Do not take yearly results too seriously. Instead, focus on four- or five-year averages.**"
> -- <cite>Benjamin Graham</cite>

::: {.content-visible when-profile="website_en"}
This quote encourages investors to adopt a long-term perspective when evaluating investments. Yearly fluctuations in a company's performance can be misleading, as they may be influenced by temporary factors. By focusing on longer-term averages, investors can gain a more accurate understanding of a company's true performance and potential, aligning with the principles of value investing.
:::

::: {.content-visible when-profile="website_fr"}
Cette citation encourage les investisseurs à adopter une perspective à long terme lorsqu'ils évaluent des investissements. Les fluctuations annuelles des performances d'une entreprise peuvent être trompeuses, car elles peuvent être influencées par des facteurs temporaires. En se concentrant sur des moyennes à plus long terme, les investisseurs peuvent obtenir une compréhension plus précise de la performance et du potentiel réels d'une entreprise, en accord avec les principes de l'investissement en valeur.
:::

---

> "**You don’t have to be smarter than the rest. You have to be more disciplined than the rest.**"
> -- <cite>Warren Buffett</cite>

::: {.content-visible when-profile="website_en"}
Buffett’s advice underscores the importance of discipline in investing. Success in value investing doesn’t require extraordinary intelligence but rather the ability to stick to a well-thought-out strategy, avoid emotional decision-making, and be patient enough to wait for the right opportunities. Discipline helps investors remain focused on their long-term goals, even in the face of market volatility.
:::

::: {.content-visible when-profile="website_fr"}
Le conseil de Buffett souligne l'importance de la discipline dans l'investissement. Le succès dans l'investissement en valeur ne nécessite pas une intelligence extraordinaire, mais plutôt la capacité à suivre une stratégie bien réfléchie, à éviter les décisions émotionnelles et à être patient en attendant les bonnes opportunités. La discipline aide les investisseurs à rester concentrés sur leurs objectifs à long terme, même face à la volatilité du marché.
:::

---

> "**Time is the friend of the wonderful business, the enemy of the mediocre.**"
> -- <cite>Warren Buffett</cite>

::: {.content-visible when-profile="website_en"}
This quote highlights the importance of time in value investing. Wonderful businesses with strong fundamentals tend to grow and create value over the long term, making time an ally for investors who choose wisely. Conversely, time can expose the weaknesses of mediocre businesses, leading to declining value. This is why value investors are patient, willing to hold onto quality companies for years.
:::

::: {.content-visible when-profile="website_fr"}
Cette citation met en lumière l'importance du temps dans l'investissement _value_. Les entreprises merveilleuses avec de solides fondamentaux ont tendance à croître et à créer de la valeur sur le long terme, faisant du temps un allié pour les investisseurs qui choisissent judicieusement. En revanche, le temps peut exposer les faiblesses des entreprises médiocres, conduisant à une baisse de la valeur. C'est pourquoi les investisseurs en _value_ sont patients et prêts à conserver des entreprises de qualité pendant des années.
:::
---

> "**The best investment you can make is in yourself.**"
> -- <cite>Warren Buffett</cite>
::: {.content-visible when-profile="website_en"}
Buffett reminds us that the most valuable investment isn’t always found in the stock market but in oneself. By continuously improving knowledge, skills, and understanding of investing principles, an individual increases their ability to make wise investment decisions. This investment in personal growth ultimately leads to better financial outcomes and greater life satisfaction.
:::

::: {.content-visible when-profile="website_fr"}
Buffett nous rappelle que l'investissement le plus précieux ne se trouve pas toujours sur le marché boursier, mais en soi. En améliorant continuellement ses connaissances, ses compétences et sa compréhension des principes d'investissement, une personne augmente sa capacité à prendre des décisions d'investissement judicieuses. Cet investissement dans la croissance personnelle conduit finalement à de meilleurs résultats financiers et à une plus grande satisfaction dans la vie.
:::
