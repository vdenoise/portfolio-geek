> Diversification is the only free lunch in investing.
> -- <cite>Harry Markowitz</cite>

::: {.content-visible when-profile="website_en"}
Harry Markowitz, the father of modern portfolio theory, revolutionized investing with the idea of diversification. By spreading investments across various asset classes—such as stocks, bonds, and real estate—investors can reduce risk without sacrificing returns. This "free lunch" refers to the ability to minimize risk by not putting all your eggs in one basket, thereby mitigating the impact of any single investment’s poor performance. Diversification works because different assets often move in opposite directions under varying market conditions, balancing the overall portfolio performance and enhancing the chances of stable returns over time.
:::

::: {.content-visible when-profile="website_fr"}
Harry Markowitz, père de la théorie moderne du portefeuille, a révolutionné l'investissement avec l'idée de la diversification. En répartissant les investissements entre différentes classes d'actifs, telles que les actions, les obligations et l'immobilier, les investisseurs peuvent réduire le risque sans sacrifier les rendements. Ce "repas gratuit" fait référence à la possibilité de minimiser les risques en ne mettant pas tous ses œufs dans le même panier, atténuant ainsi l'impact de la mauvaise performance d'un investissement particulier. La diversification fonctionne car différents actifs évoluent souvent dans des directions opposées selon les conditions du marché, équilibrant la performance globale du portefeuille et augmentant les chances de rendements stables sur le long terme.
:::

> The essence of investment management is the management of risks, not the management of returns.
> -- <cite>Benjamin Graham</cite>

::: {.content-visible when-profile="website_en"}
Benjamin Graham, known as the father of value investing, believed that focusing on risk management is paramount. Returns are uncertain and largely dependent on market conditions, but risk is something investors can control by carefully selecting assets, diversifying, and being cautious of overpaying for securities. Managing risk ensures that the portfolio can withstand market downturns and avoid catastrophic losses. By prioritizing risk management, investors protect their capital, making it possible to achieve long-term investment goals even if short-term returns are unpredictable or lower than expected.
:::


::: {.content-visible when-profile="website_fr"}
Benjamin Graham, connu comme le père de l'investissement en valeur, croyait que la gestion des risques est primordiale. Les rendements sont incertains et largement dépendants des conditions du marché, mais le risque est quelque chose que les investisseurs peuvent contrôler en sélectionnant soigneusement les actifs, en diversifiant et en étant prudents quant au prix payé pour les titres. La gestion des risques garantit que le portefeuille peut résister aux baisses du marché et éviter des pertes catastrophiques. En priorisant la gestion des risques, les investisseurs protègent leur capital, ce qui leur permet d'atteindre des objectifs d'investissement à long terme, même si les rendements à court terme sont imprévisibles ou inférieurs aux attentes.
:::


---

> The goal of a successful portfolio is to have a balance of risk and return, tailored to your unique goals and risk tolerance.
> -- <cite>Suze Orman</cite>

::: {.content-visible when-profile="website_en"}
Suze Orman emphasizes the importance of aligning your investment strategy with personal financial goals and risk tolerance. A well-constructed portfolio balances growth potential and safety, ensuring that the investor can achieve their objectives—be it retirement, purchasing a home, or funding education—without taking on more risk than they are comfortable with. This balance requires understanding both the time horizon for each goal and the investor’s emotional capacity to handle market volatility. By customizing the portfolio to reflect these factors, investors can maintain confidence and stay the course through market ups and downs.
:::

::: {.content-visible when-profile="website_fr"}
Suze Orman souligne l'importance d'aligner sa stratégie d'investissement avec ses objectifs financiers personnels et sa tolérance au risque. Un portefeuille bien construit équilibre le potentiel de croissance et la sécurité, garantissant que l'investisseur puisse atteindre ses objectifs, qu'il s'agisse de la retraite, de l'achat d'une maison ou du financement de l'éducation, sans prendre plus de risques qu'il n'en est capable. Cet équilibre nécessite de comprendre à la fois l'horizon temporel pour chaque objectif et la capacité émotionnelle de l'investisseur à gérer la volatilité du marché. En personnalisant le portefeuille pour refléter ces facteurs, les investisseurs peuvent maintenir leur confiance et rester sur la bonne voie à travers les hauts et les bas du marché.
:::

---



> A portfolio that is not well-diversified exposes you to unnecessary risks and limits your potential for growth.
> -- <cite>David Swensen</cite>

::: {.content-visible when-profile="website_en"}
David Swensen, renowned for managing Yale’s endowment, advocated for diversification as a key strategy in reducing risk. A concentrated portfolio—where investments are limited to a few assets or sectors—can suffer severe losses if those areas underperform. By diversifying across asset classes, sectors, and geographic regions, investors can mitigate the risk of any single investment causing significant harm to their portfolio. This broader exposure not only protects against downside risk but also captures opportunities for growth across different areas, enhancing overall portfolio performance in various market conditions.
:::

::: {.content-visible when-profile="website_fr"}
David Swensen, célèbre pour avoir géré le fonds de dotation de Yale, a plaidé pour la diversification comme stratégie clé pour réduire les risques. Un portefeuille concentré, où les investissements sont limités à quelques actifs ou secteurs, peut subir des pertes importantes si ces secteurs sous-performent. En diversifiant les investissements entre différentes classes d'actifs, secteurs et régions géographiques, les investisseurs peuvent atténuer le risque qu'un seul investissement cause des dommages significatifs à leur portefeuille. Cette exposition plus large protège non seulement contre les risques de baisse, mais capture également les opportunités de croissance dans différents domaines, améliorant la performance globale du portefeuille dans diverses conditions de marché.
:::

---

> Don’t look for the needle in the haystack. Just buy the haystack!
> -- <cite>John C. Bogle</cite>

::: {.content-visible when-profile="website_en"}
John C. Bogle, founder of Vanguard and the pioneer of index funds, advocated for broad market exposure through index investing rather than trying to pick individual stocks (the needle). Index funds represent the "haystack," capturing the performance of entire markets or sectors, thus reducing the risk associated with selecting individual winners. This approach is rooted in the belief that it’s nearly impossible to consistently outperform the market through stock picking. By buying the haystack, investors can benefit from the overall growth of the market, ensuring a diversified and low-cost investment approach with steady long-term returns.
:::

::: {.content-visible when-profile="website_fr"}
John C. Bogle, fondateur de Vanguard et pionnier des fonds indiciels, a préconisé une exposition large au marché à travers l'investissement indiciel plutôt que d'essayer de sélectionner des actions individuelles (l'aiguille). Les fonds indiciels représentent la "meule de foin", capturant la performance de l'ensemble des marchés ou des secteurs, réduisant ainsi le risque associé à la sélection des gagnants individuels. Cette approche repose sur la conviction qu'il est presque impossible de surperformer le marché de manière cohérente en sélectionnant des actions. En achetant la meule de foin, les investisseurs peuvent bénéficier de la croissance globale du marché, assurant une approche d'investissement diversifiée et à faible coût avec des rendements stables à long terme.
:::

---

> Your portfolio should be like a well-balanced meal, with a mix of growth, income, and stability to sustain you over the long term.
> -- <cite>Christine Benz</cite>

::: {.content-visible when-profile="website_en"}
Christine Benz uses the metaphor of a well-balanced meal to describe a diversified portfolio. Just as a nutritious diet includes a variety of food groups, a robust portfolio should contain different types of investments to meet long-term financial needs. Growth assets (like stocks) offer the potential for capital appreciation, income assets (such as bonds) provide regular interest payments, and stability-focused investments (like cash or conservative bonds) offer protection against volatility. This mix helps investors achieve a balanced approach that supports sustained growth while managing risk, much like how a balanced diet promotes long-term health.
:::

::: {.content-visible when-profile="website_fr"}
Christine Benz utilise la métaphore d'un repas équilibré pour décrire un portefeuille diversifié. Tout comme une alimentation nutritive comprend une variété de groupes alimentaires, un portefeuille robuste doit contenir différents types d'investissements pour répondre aux besoins financiers à long terme. Les actifs de croissance (comme les actions) offrent un potentiel d'appréciation du capital, les actifs générateurs de revenus (tels que les obligations) fournissent des paiements d'intérêts réguliers, et les investissements axés sur la stabilité (comme le cash ou les obligations conservatrices) offrent une protection contre la volatilité. Ce mélange aide les investisseurs à atteindre une approche équilibrée qui favorise une croissance soutenue tout en gérant les risques, tout comme une alimentation équilibrée favorise la santé à long terme.
:::
---

> Successful investing is about managing risk, not avoiding it.
> -- <cite>Benjamin Graham</cite>

::: {.content-visible when-profile="website_en"}
Benjamin Graham's wisdom highlights the importance of acknowledging and managing risk rather than attempting to avoid it altogether, which is often unrealistic in investing. All investments carry some degree of risk, but by understanding and controlling that risk—through diversification, proper asset allocation, and thorough analysis—investors can protect themselves from significant losses. This approach allows investors to stay invested through market fluctuations and capitalize on opportunities without being derailed by fear or volatility, thereby increasing the likelihood of long-term success and reaching their financial goals.
:::

::: {.content-visible when-profile="website_fr"}
La perspective de Benjamin Graham souligne l'importance de reconnaître et de gérer le risque plutôt que de tenter de l'éviter complètement, ce qui est souvent irréaliste en matière d'investissement. Tous les investissements comportent un certain degré de risque, mais en comprenant et en contrôlant ce risque, par exemple en diversifiant, en répartissant correctement les actifs et en réalisant des analyses approfondies, les investisseurs peuvent se protéger contre des pertes significatives. Cette approche permet aux investisseurs de rester investis pendant les fluctuations du marché et de capitaliser sur les opportunités sans être perturbés par la peur ou la volatilité, augmentant ainsi les chances de succès à long terme et d'atteinte de leurs objectifs financiers.
:::

---

> In portfolio construction, the objective is not to maximize returns, but to maximize the likelihood of achieving your long-term goals.
> -- <cite>William Bernstein</cite>

::: {.content-visible when-profile="website_en"}
William Bernstein, a financial theorist and author, stresses that portfolio construction should focus on achieving long-term financial goals rather than chasing maximum returns. This involves designing a portfolio that balances growth potential with risk management, ensuring that it aligns with the investor’s specific objectives, such as retirement or education funding. By focusing on goals rather than returns, investors are more likely to adopt a disciplined, patient approach, avoiding the pitfalls of speculative investing or market timing. This strategy increases the likelihood of reaching financial targets without unnecessary exposure to risk.
:::

::: {.content-visible when-profile="website_fr"}
William Bernstein, théoricien financier et auteur, souligne que la construction de portefeuille doit se concentrer sur l'atteinte des objectifs financiers à long terme plutôt que sur la recherche des rendements maximaux. Cela implique de concevoir un portefeuille qui équilibre le potentiel de croissance et la gestion des risques, en s'assurant qu'il s'aligne sur les objectifs spécifiques de l'investisseur, tels que la retraite ou le financement de l'éducation. En se concentrant sur les objectifs plutôt que sur les rendements, les investisseurs sont plus susceptibles d'adopter une approche disciplinée et patiente, en évitant les pièges de l'investissement spéculatif ou du timing du marché. Cette stratégie augmente les chances d'atteindre les objectifs financiers sans s'exposer à des risques inutiles.
:::

> By accepting the reality of market uncertainty, a diversified portfolio allows you to weather storms and benefit from long-term growth.
> -- <cite>Meb Faber</cite>

::: {.content-visible when-profile="website_en"}
Meb Faber, an investment strategist, emphasizes the importance of accepting market uncertainty as a constant and using diversification as a strategy to manage it. Markets are inherently unpredictable, and trying to time them can lead to poor decisions. A diversified portfolio, however, spreads investments across different assets, reducing the impact of any one market downturn. This approach not only protects the portfolio during volatile periods but also positions it to benefit from long-term growth as markets recover and different assets perform well at different times, leading to more stable overall returns.
:::

::: {.content-visible when-profile="website_fr"}
Meb Faber, stratège en investissement, met l'accent sur l'importance d'accepter l'incertitude du marché comme une constante et d'utiliser la diversification comme stratégie pour la gérer. Les marchés sont par nature imprévisibles, et essayer de les synchroniser peut conduire à de mauvaises décisions. Un portefeuille diversifié, cependant, répartit les investissements entre différents actifs, réduisant l'impact de la baisse d'un marché spécifique. Cette approche non seulement protège le portefeuille pendant les périodes de volatilité, mais le positionne également pour bénéficier de la croissance à long terme lorsque les marchés se redressent et que différents actifs performent bien à différents moments, conduisant à des rendements globaux plus stables.
