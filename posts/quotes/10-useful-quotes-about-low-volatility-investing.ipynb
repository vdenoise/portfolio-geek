> **The challenge is to find the right balance between risk and return.**
> -- <cite>Harry Markowitz</cite>

::: {.content-visible when-profile="website_en"}
Harry Markowitz, known for his Modern Portfolio Theory, underscores the balance required in investing. In low volatility investing, the aim is to achieve a favourable risk-return profile by choosing stocks that exhibit lower volatility. This means that while the potential returns might not be as high as with more volatile investments, the risk is also lower. The concept of risk-adjusted return becomes central here, as investors seek to maximise returns while minimising risk. Markowitz’s theory emphasises that a well-diversified portfolio with low-volatility stocks can achieve this balance, leading to a more stable and potentially more rewarding investment experience.
:::

::: {.content-visible when-profile="website_fr"}
Harry Markowitz, connu pour sa Théorie Moderne du Portefeuille, souligne l'équilibre nécessaire dans l'investissement. Dans l'investissement à faible volatilité, l'objectif est d'atteindre un profil risque-rendement favorable en choisissant des actions présentant une volatilité plus faible. Cela signifie que, bien que les rendements potentiels puissent ne pas être aussi élevés que ceux des investissements plus volatils, le risque est également réduit. Le concept de rendement ajusté au risque devient central ici, car les investisseurs cherchent à maximiser les rendements tout en minimisant le risque. La théorie de Markowitz souligne qu'un portefeuille bien diversifié avec des actions à faible volatilité peut atteindre cet équilibre, conduisant à une expérience d'investissement plus stable et potentiellement plus gratifiante.
