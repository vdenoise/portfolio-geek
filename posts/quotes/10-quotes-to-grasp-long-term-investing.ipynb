> **"The individual investor should act consistently as an investor and not as a speculator."**  
> -- <cite>Benjamin Graham</cite>  

::: {.content-visible when-profile="website_en"}  
Benjamin Graham, known as the father of value investing, differentiates between investing and speculating. Investing involves making decisions based on careful analysis, focusing on the long-term potential of an asset. Speculation, on the other hand, often involves taking on excessive risk in hopes of quick profits. Graham's advice to act as an investor rather than a speculator is a foundational principle for long-term investing success. By maintaining a disciplined approach and focusing on value, investors can avoid the pitfalls of speculation and build sustainable wealth over time.  
:::

::: {.content-visible when-profile="website_fr"}  
Benjamin Graham, connu comme le père de l'investissement dans la valeur, fait la distinction entre investir et spéculer. Investir consiste à prendre des décisions basées sur une analyse minutieuse, en se concentrant sur le potentiel à long terme d'un actif. La spéculation, quant à elle, implique souvent de prendre des risques excessifs dans l'espoir de gains rapides. Le conseil de Graham de se comporter en investisseur plutôt qu'en spéculateur est un principe fondamental pour réussir dans l'investissement à long terme. En maintenant une approche disciplinée et en se concentrant sur la valeur, les investisseurs peuvent éviter les pièges de la spéculation et accumuler une richesse durable au fil du temps.
