## 2. Key Concepts and Formulas

---

### 2.1. The Formula for Historical Volatility
Historical volatility is calculated using the standard deviation of logarithmic returns of an asset's prices.

$$ \sigma = \sqrt{\frac{1}{N-1} \sum_{i=1}^{N} (R_i - \bar{R})^2} $$

where:
- \( \sigma \) is the historical volatility
- \( N \) is the number of observations
- \( R_i \) is the logarithmic return at time \( i \)
- \( \bar{R} \) is the average of the logarithmic returns

### 2.2. Annualizing Historical Volatility
To annualize the calculated weekly volatility, use the following formula:

$$ \sigma_{\text{annual}} = \sigma_{\text{weekly}} \times \sqrt{52} $$

### 2.3. Parameters and Their Impact
The number of observations (N), the time interval (weekly), and the asset's historical closing prices significantly influence the calculated volatility. Understanding these parameters helps in tweaking the analysis for different assets or time frames.

## 3. Practical Implementation in Python

---

### 3.1. Importing Required Libraries
First, let's import the necessary libraries for our analysis.


```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 3.2. Fetching Historical Data
We'll use the `yfinance` library to fetch historical data for SPY (S&P 500 ETF).


```python
# Fetch SPY data
spy = yf.Ticker('SPY')
df = spy.history(period='5y', interval='1wk')
df.head()
```

### 3.3. Calculating Weekly Returns
Using the historical weekly closing prices, we calculate the logarithmic returns.


```python
# Calculate logarithmic returns
df['Log Returns'] = np.log(df['Close'] / df['Close'].shift(1))
df = df.dropna()
df.head()
```

### 3.4. Calculating Historical Volatility
Next, we compute the historical volatility using the standard deviation of the logarithmic returns and annualize it.


```python
# Calculate historical volatility
weekly_volatility = df['Log Returns'].std()
annual_volatility = weekly_volatility * np.sqrt(52)
print(f'Annualized Historical Volatility: {annual_volatility:.2%}')
```

### 3.5. Visualising Volatility
We'll plot the weekly returns and volatility to visualise the data better.


```python
# Plot weekly returns
df['Log Returns'].plot(figsize=(14, 7), title='Weekly Log Returns of SPY')
plt.show()
```
