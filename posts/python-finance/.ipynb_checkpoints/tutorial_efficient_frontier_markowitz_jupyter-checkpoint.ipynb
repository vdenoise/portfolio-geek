# Libraries: setup and import

```python
%%capture
!pip install yfinance PyPortfolioOpt pandas numpy matplotlib

import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import expected_returns, risk_models
```

# Define the investment universe

```python
tickers = [
    "SPY",   # US Equities (S&P 500)
    "GOVT",  # US Government Bonds
    "LQD",   # US Corporate Bonds
    "HYG",   # US High Yield Bonds
    "VGK",   # European Equities
    "EWJ",   # Japan Equities
    "GLD",   # Gold
    "GSG",   # Commodities
    "VNQ",   # Real Estate (REIT)
    "EMB",   # Emerging Market Bonds
]
```

## Get Data

```python
data = yf.download(tickers, start="2020-01-01", end="2025-01-01", progress=False)
data = data["Close"]  # Only keep the 'Close' prices
data.dropna(how="all", inplace=True)  # Remove rows with all NaN values
```

```python
data.tail(3)
```

# Get Key Stats for the Model

```python
mu = expected_returns.mean_historical_return(data[tickers], frequency=252)
Sigma = risk_models.sample_cov(data[tickers], frequency=252)
```

```python
mu
```

# Build the Efficient Frontier 1/2

## Efficient Frontier and Key Portfolios

```python
# 5) Build the Efficient Frontier and portfolio weights
ef = EfficientFrontier(mu, Sigma, weight_bounds=(0, 1))

# Minimum volatility portfolio
min_vol_weights = ef.min_volatility()
ret_min_vol, vol_min_vol, _ = ef.portfolio_performance()

# Maximum return portfolio
ef = EfficientFrontier(mu, Sigma, weight_bounds=(0, 1))
ef.max_quadratic_utility()
ret_max_ret, vol_max_ret, _ = ef.portfolio_performance()

# Maximum Sharpe portfolio
ef = EfficientFrontier(mu, Sigma, weight_bounds=(0, 1))
max_sharpe_weights = ef.max_sharpe()
ret_tan, vol_tan, _ = ef.portfolio_performance()
```

# Build the Efficient Frontier 2/2

## Calculating the Full Coordinates for the Efficient Frontier

```python
# Efficient Frontier data
target_returns = np.linspace(ret_min_vol, ret_max_ret, 50)
frontier_vols = []
frontier_rets = []

for tr in target_returns:
    ef_temp = EfficientFrontier(mu, Sigma, weight_bounds=(0, 1))
    try:
        ef_temp.efficient_return(target_return=tr)
        ret_p, vol_p, _ = ef_temp.portfolio_performance()
        frontier_rets.append(ret_p)
        frontier_vols.append(vol_p)
    except:
        pass
```

# Printing Results

```python
print("=== Portfolio Weights ===")
print("\nMinimum Volatility Portfolio:")
for tck, w in min_vol_weights.items():
    if w > 1e-4:
        print(f"{tck}: {w:.2%}")
print(f"Return: {ret_min_vol*100:.2f}%, Volatility: {vol_min_vol*100:.2f}%")

print("\nMaximum Sharpe Portfolio:")
for tck, w in max_sharpe_weights.items():
    if w > 1e-4:
        print(f"{tck}: {w:.2%}")
print(f"Return: {ret_tan*100:.2f}%, Volatility: {vol_tan*100:.2f}%")
```

# Plotting the Efficient Frontier

```python
fig, ax = plt.subplots(figsize=(12, 8))
asset_vols = np.sqrt(np.diag(Sigma))
ax.scatter(asset_vols, mu, c='red', marker='o', s=100, label='Individual ETFs')
for i, txt in enumerate(tickers):
    ax.text(asset_vols[i] + 0.005, mu[i], txt, fontsize=9, ha='left', va='center')
ax.plot(frontier_vols, frontier_rets, 'b--', linewidth=2, label='Efficient Frontier');
```

## Adding Max Sharpe, Min Volatility and Max Return Portfolio

```python
fig, ax = plt.subplots(figsize=(12, 8))
asset_vols = np.sqrt(np.diag(Sigma))
ax.scatter(asset_vols, mu, c='red', marker='o', s=100, label='Individual ETFs')
for i, txt in enumerate(tickers):
    ax.text(asset_vols[i] + 0.005, mu[i], txt, fontsize=9, ha='left', va='center')
ax.plot(frontier_vols, frontier_rets, 'b--', linewidth=2, label='Efficient Frontier');

asset_vols = np.sqrt(np.diag(Sigma))
ax.scatter(asset_vols, mu, c='red', marker='o', s=100, label='Individual ETFs')
for i, txt in enumerate(tickers):
    ax.text(asset_vols[i] + 0.005, mu[i], txt, fontsize=9, ha='left', va='center')
ax.plot(frontier_vols, frontier_rets, 'b--', linewidth=2, label='Efficient Frontier');

ax.plot(frontier_vols, frontier_rets, 'b--', linewidth=2, label='Efficient Frontier')
# Minimum Volatility Portfolio
ax.scatter([vol_min_vol], [ret_min_vol], c='blue', marker='o', s=150, label='Min Vol Portfolio')
ax.text(vol_min_vol + 0.01, ret_min_vol, f"Min Vol\n{vol_min_vol*100:.1f}%, {ret_min_vol*100:.1f}%", fontsize=9)

# Maximum Sharpe Portfolio
ax.scatter([vol_tan], [ret_tan], c='green', marker='*', s=300, label='Max Sharpe Portfolio')
ax.text(vol_tan + 0.01, ret_tan, f"Max Sharpe\n{vol_tan*100:.1f}%, {ret_tan*100:.1f}%", fontsize=9)

# Maximum Return Portfolio
ax.scatter([vol_max_ret], [ret_max_ret], c='orange', marker='o', s=150, label='Max Return Portfolio')
ax.text(vol_max_ret + 0.01, ret_max_ret, f"Max Ret\n{vol_max_ret*100:.1f}%, {ret_max_ret*100:.1f}%", fontsize=9)
plt.show()
```

# Adding the Capital Market Line

```python
risk_free_rate = 0.04  # 4% per annum
fig, ax = plt.subplots(figsize=(12, 8))
asset_vols = np.sqrt(np.diag(Sigma))
ax.scatter(asset_vols, mu, c='red', marker='o', s=100, label='Individual ETFs')
for i, txt in enumerate(tickers):
    ax.text(asset_vols[i] + 0.005, mu[i], txt, fontsize=9, ha='left', va='center')
ax.plot(frontier_vols, frontier_rets, 'b--', linewidth=2, label='Efficient Frontier');

asset_vols = np.sqrt(np.diag(Sigma))
ax.scatter(asset_vols, mu, c='red', marker='o', s=100, label='Individual ETFs')
for i, txt in enumerate(tickers):
    ax.text(asset_vols[i] + 0.005, mu[i], txt, fontsize=9, ha='left', va='center')
ax.plot(frontier_vols, frontier_rets, 'b--', linewidth=2, label='Efficient Frontier');

ax.plot(frontier_vols, frontier_rets, 'b--', linewidth=2, label='Efficient Frontier')
# Minimum Volatility Portfolio
ax.scatter([vol_min_vol], [ret_min_vol], c='blue', marker='o', s=150, label='Min Vol Portfolio')
ax.text(vol_min_vol + 0.01, ret_min_vol, f"Min Vol\n{vol_min_vol*100:.1f}%, {ret_min_vol*100:.1f}%", fontsize=9)

# Maximum Sharpe Portfolio
ax.scatter([vol_tan], [ret_tan], c='green', marker='*', s=300, label='Max Sharpe Portfolio')
ax.text(vol_tan + 0.01, ret_tan, f"Max Sharpe\n{vol_tan*100:.1f}%, {ret_tan*100:.1f}%", fontsize=9)

# Maximum Return Portfolio
ax.scatter([vol_max_ret], [ret_max_ret], c='orange', marker='o', s=150, label='Max Return Portfolio')
ax.text(vol_max_ret + 0.01, ret_max_ret, f"Max Ret\n{vol_max_ret*100:.1f}%, {ret_max_ret*100:.1f}%", fontsize=9)

# Capital Market Line (CML)
cml_x = np.linspace(0, vol_tan * 1.2, 100)
cml_y = risk_free_rate + ((ret_tan - risk_free_rate) / vol_tan) * cml_x
plt.plot(cml_x, cml_y, color='green', linewidth=2, label='Capital Market Line')

# Formatting
plt.title("Efficient Frontier & Capital Market Line with Portfolio Weights", fontsize=16, fontweight='bold', pad=20)
plt.xlabel("Annualized Volatility (Risk) [%]", fontsize=12, labelpad=10)
plt.ylabel("Annualized Return [%]", fontsize=12, labelpad=10)
plt.legend(loc='best', fontsize=12)
plt.grid(alpha=0.4)

# Format the x and y axis values as percentages
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{x*100:.0f}%"))
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{x*100:.0f}%"))

# Show the plot
plt.tight_layout()
plt.show()
```
